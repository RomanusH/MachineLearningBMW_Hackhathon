```python
# 1. Replace missing values with 0
data_zero = data.copy()
data_zero.fillna(0, inplace=True)

# 2. Leave the missing values as they are
data_as_is = data.copy()
```

```python
# 1. Replace missing values with the mode
data_mode = data.copy()
for col in data_mode.select_dtypes(include=['number']).columns:  # Only numeric columns
    data_mode[col].fillna(data_mode[col].mode()[0], inplace=True)

# 2. Replace missing values with the median
data_median = data.copy()
data_median.fillna(data_median.median(), inplace=True)

# 3. Replace missing values with the mean
data_mean = data.copy()
data_mean.fillna(data_mean.mean(), inplace=True)
```
